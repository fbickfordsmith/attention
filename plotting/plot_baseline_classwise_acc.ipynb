```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/Users/fbickfordsmith/Google Drive/Project/'
df = pd.read_csv(path+'attention/results/baseline_classwise_acc.csv', index_col=0)
sorted_df = df.sort_values(by='accuracy', ascending=False)
```

```python
fig, ax = plt.subplots()
plt.plot(range(1000), sorted_df['accuracy'])
plt.xlabel('Class')
# plt.tick_params(
#     axis='x',          # changes apply to the x-axis
#     which='both',      # both major and minor ticks are affected
#     bottom=False,      # ticks along the bottom edge are off
#     top=False,         # ticks along the top edge are off
#     labelbottom=False) # labels along the bottom edge are off
plt.xticks((0, 999))
ax.set_xticklabels((1, 1000))
plt.ylabel('Top-1 accuracy')
# plt.savefig('baseline_classwise_acc.png', dpi=600)
plt.show()
# plt.clf()
```

```python
mean = np.mean(df['accuracy'])
std = np.std(df['accuracy'])
classes_within_1std = np.flatnonzero((df['accuracy']>mean-std) & (df['accuracy']<mean+std))

fig, ax = plt.subplots()
ax.hist(df['accuracy'], bins=15, color='w', edgecolor='black') #alpha=0.5)
ax.axvline(mean, color='k', linestyle='--')
# ax.axvline(mean-std, color='r', linestyle='--')
# ax.axvline(mean+std, color='r', linestyle='--')
plt.xlim((0, 1))
plt.xlabel('Baseline accuracy')
plt.ylabel('Number of classes')
# plt.savefig('baseline_classwise_acc_hist.png', dpi=600)
plt.show()
```
