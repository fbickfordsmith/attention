```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/Users/fbickfordsmith/Google Drive/Project/attention/'
```

```python
plt.style.use('seaborn-bright')
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
linestyles = ['-' for _ in range(len(colors))]
linestyles.extend('--' for _ in range(len(colors)))
```

### All training vs all validation

```python
type_context = 'size'

fig, ax = plt.subplots(1, 2, sharey=True, figsize=(12, 4))

for i in range(9):
    df = pd.read_csv(f'{path}training/{type_context}context{i:02}_training.csv', index_col=0)
    ax[0].plot(df['loss'], label=i+1, linestyle=linestyles[i])
    ax[1].plot(df['val_loss'], label=i+1, linestyle=linestyles[i])
    
ax[0].set_xlabel('Epoch')
ax[1].set_xlabel('Epoch')
ax[0].set_ylabel('Loss (training set)')
ax[1].set_ylabel('Loss (validation set)')
ax[0].legend()
ax[1].legend()
plt.tight_layout()
plt.savefig(f'{path}plotting/sizecontexts_training.png', dpi=200)
plt.show()
```

```python
type_context = 'size'

fig, ax = plt.subplots(1, 2, sharey=True, figsize=(12, 4))

for i in range(9):
    df = pd.read_csv(f'{path}training/{type_context}context{i:02}_training.csv', index_col=0)
    ax[0].plot(df['acc'], label=i+1, linestyle=linestyles[i])
    ax[1].plot(df['val_acc'], label=i+1, linestyle=linestyles[i])
    
ax[0].set_xlabel('Epoch')
ax[1].set_xlabel('Epoch')
ax[0].set_ylabel('Accuracy (training set)')
ax[1].set_ylabel('Accuracy (validation set)')
ax[0].legend()
ax[1].legend()
plt.tight_layout()
# plt.savefig(f'{path}plotting/sizecontexts_training.png', dpi=200)
plt.show()
```

### Model by model

```python
type_context = 'size'
shape_fig = (3, 3)

fig, ax = plt.subplots(*shape_fig, sharex=True, sharey=True, figsize=(10, 6))

for i in range(9):
    df = pd.read_csv(f'{path}training/{type_context}context{i:02}_training.csv', index_col=0)
    rowcol = np.unravel_index(i, shape_fig)
    ax[rowcol].plot(df['loss'], label='training')
    ax[rowcol].plot(df['val_loss'], label='validation')
    ax[rowcol].set_title(f'Size context {i+1}')
    ax[rowcol].legend()
    if rowcol[1] == 0:
        ax[rowcol].set_ylabel('Loss')
    if rowcol[0] == 2:
        ax[rowcol].set_xlabel('Epoch')

plt.tight_layout()
plt.savefig(f'{path}plotting/sizecontexts_training_loss.png', dpi=200)
plt.show()
```
