```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vgg16/attention-model-results.csv', index_col=0)
df
```

```python
plt.plot(df['acc'], label='training')
plt.plot(df['val_acc'], label='validation')
plt.xlabel('epoch')
plt.ylabel('top-1 accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(df['top_k_categorical_accuracy'], label='training')
plt.plot(df['val_top_k_categorical_accuracy'], label='validation')
plt.xlabel('epoch')
plt.ylabel('top-5 accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(df['loss'], label='training')
plt.plot(df['val_loss'], label='validation')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.show()
```
