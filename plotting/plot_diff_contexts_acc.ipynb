```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from  sklearn.linear_model import LinearRegression
```

```python
path = '/Users/fbickfordsmith/Google Drive/Project/'
df0 = pd.read_csv(path+'attention/contexts/diffcontexts_stats.csv', index_col=0)
df0
```

```python
df1 = pd.read_csv(path+'attention/results/diffcontexts_trained_metrics.csv', index_col=0)
df1
```

```python
# plt.figure(figsize=(10, 6))
fig, ax = plt.subplots()
plt.plot(df0['incontext_acc'], '-', label='in context (base)')
plt.plot(df0['outofcontext_acc'], '-', label='out of context (base)')
plt.plot(df1['incontext_acc'], '--', label='in context (attention)')
plt.plot(df1['outofcontext_acc'], '--', label='out of context (attention)')
# plt.plot(df1['alldata_acc'], '--', label='all data (attention)')
# plt.plot(range(20), [np.sum(df0['num_correct'])/np.sum(df0['num_examples']) for _ in range(20)])
plt.xlabel('Training context')
plt.ylabel('Top-1 accuracy')
plt.xticks((0, 19))
ax.set_xticklabels((1, 20))
plt.legend(title='Source of test examples')
# plt.savefig('diff_contexts_acc.png', dpi=600)
plt.show()
```

```python
fig, ax = plt.subplots()
change_incontext = df1['incontext_acc']-df0['incontext_acc']
change_outofcontext = df1['outofcontext_acc']-df0['outofcontext_acc']
x = np.arange(20)
w = 0.4
plt.bar(x-w/2, change_incontext, width=w, label='in context')
plt.bar(x+w/2, change_outofcontext, width=w, label='out of context')
plt.xlabel('Training context')
plt.ylabel('Change in top-1 accuracy')
plt.ylim((-0.01, 0.09))
plt.xticks((0, 19))
# plt.yticks(range(-0.01, 0.09, 0.01))
ax.set_xticklabels((1, 20))
plt.legend(title='Source of test examples')
# plt.savefig('diff_contexts_acc_change.png', dpi=600)
plt.show()
```

```python
x = np.array(df0['incontext_acc'])
y = df1['incontext_acc']
d = 0.02
plt.plot(x, y)
plt.plot((min(x)-d, max(x)+d), (min(x)-d, max(x)+d), '--')
plt.show()
```

```python
x = 1 - np.array(df0['incontext_acc'])
plt.plot(x, df0['incontext_acc'])
plt.plot(x, df1['incontext_acc'])
plt.plot(x, df0['outofcontext_acc'])
plt.plot(x, df1['outofcontext_acc'])
plt.xlabel('Difficulty of training context\n'
           '(in-context error rate of base model)')
plt.ylabel('Accuracy on in-context examples')
plt.show()
```
