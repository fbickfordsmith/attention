```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/Users/fbickfordsmith/Google Drive/Project/attention/'
```

```python
plt.style.use('seaborn-bright')
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
linestyles = ['-' for _ in range(len(colors))]
linestyles.extend('--' for _ in range(len(colors)))
```

```python
type_context = 'sem'
shape_fig = (3, 2)

fig, ax = plt.subplots(*shape_fig, sharex=True, sharey=False, figsize=(10, 6))

for i in range(np.prod(shape_fig)):
    df = pd.read_csv(f'{path}training/{type_context}context{i:02}_training_v8.csv', index_col=0)
    rowcol = np.unravel_index(i, shape_fig)
    ax[rowcol].plot(df['loss'], label='training')
    ax[rowcol].plot(df['val_loss'], label='validation')
    ax[rowcol].set_title(f'{type_context}context{i+1:02}')
    ax[rowcol].legend()
    if rowcol[1] == 0:
        ax[rowcol].set_ylabel('Loss')
    if rowcol[0] == 2:
        ax[rowcol].set_xlabel('Epoch')

# plt.ylim((0.25, 1.25))
plt.tight_layout()
# plt.savefig(f'{path}plotting/sizecontexts_training_loss_v4.png', dpi=200)
plt.show()
```
