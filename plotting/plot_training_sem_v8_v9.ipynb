```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/Users/fbickfordsmith/Google Drive/Project/attention/training/'
```

```python
plt.style.use('seaborn-bright')
```

```python
version_context = 'sem_v8'
shape_fig = (3, 2)

fig, ax = plt.subplots(*shape_fig, sharex=True, sharey=False, figsize=(10, 6))

for i in range(np.prod(shape_fig)):
    df = pd.read_csv(f'{path}{version_context}_{i:02}_training.csv', index_col=0)
    rowcol = np.unravel_index(i, shape_fig)
    ax[rowcol].plot(df['loss'], label='training')
    ax[rowcol].plot(df['val_loss'], label='validation')
    ax[rowcol].set_title(f'{version_context}_{i+1:02}')
    ax[rowcol].legend()
    if rowcol[1] == 0:
        ax[rowcol].set_ylabel('Loss')
    if rowcol[0] == 2:
        ax[rowcol].set_xlabel('Epoch')

# plt.ylim((0.25, 1.25))
plt.tight_layout()
plt.show()
```
